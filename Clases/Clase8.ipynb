# Reactor Batch

```python
### version con matplotlib, es mas lento que plotly pero es mas facil de editar
def model_batch(t, x, k):
    ca0, cb0, cc0 = x
    k1, k2 = k
    ca = ca0 * np.exp(-k1 * t)
    cb = k1 * ca0 / (k2 - k1) * (np.exp(-k1 * t) - np.exp(-k2 * t))
    cc = ca0 / (k2 - k1) * (k2 * (1-np.exp(-k1 * t)) - k1 * (1-np.exp( -k2 * t)))
    return [ca, cb, cc]


def update_plots(ca0, cb0, cc0, k1, k2):
    try:
        t_span = (0, 25)
        t_eval = np.arange(*t_span, 0.01)
        x0 = [ca0, cb0, cc0]
        k = [k1, k2]
        ca, cb, cc = model_batch(t_eval, x0, k)
        
        max_cb = np.argmax(cb)

        plt.figure(figsize=(10, 6))
        plt.plot(t_eval, ca, label='Ca(t)')
        plt.plot(t_eval, cb, label='Cb(t)')
        plt.plot(t_eval, cc, label='Cc(t)')
        plt.axvline(t_eval[max_cb], color='r', linestyle='--')
        plt.text(t_eval[max_cb], 1.5, (f'Max Cb: {cb[max_cb]:.2f} at t={t_eval[max_cb]:.2f}\n'+
                                    f'Ca={ca[max_cb]:.2f}, Cb={cb[max_cb]:.2f}, Cc={cc[max_cb]:.2f}'+
                                    f'\nS_dc = {cb[max_cb]/cc[max_cb]:.2f}, Yb = {cb[max_cb]/(ca0 - ca[max_cb]):.2f}'), 
                color='black', fontsize=10, ha='left', va='bottom',
                bbox=dict(facecolor='white', alpha=1, edgecolor='black'))
        plt.xlabel('Time')
        plt.ylabel('Concentration')
        plt.title('Batch Reactor Simulation')
        plt.legend()
        plt.grid()
        plt.show()
    except Exception as e:
        print(f"Error: {e}")
# Create interactive widgets

ca0_slider = widgets.FloatSlider(value=2, min=0.1, max=5, step=0.1, description='Ca0:')
cb0_slider = widgets.FloatSlider(value=0, min=0.1, max=5, step=0.1, description='Cb0:')
cc0_slider = widgets.FloatSlider(value=0, min=0.1, max=5, step=0.1, description='Cc0:')
k1_slider = widgets.FloatSlider(value=0.5, min=0, max=1, step=0.1, description='k1:')
k2_slider = widgets.FloatSlider(value=0.2, min=0, max=1, step=0.1, description='k2:')

widgets.interact(update_plots, ca0 = ca0_slider, cb0 = cb0_slider, cc0 = cc0_slider, 
                 k1 = k1_slider, k2 = k2_slider)
```

# CSTR

```python
# 
def model_cstr(tau, x, k):
    ca0, cb0, cc0 = x
    k1, k2 = k
    ca = ca0 / (1 + k1 * tau)
    cb = (k1 * ca0 * tau) / ((1 + k1* tau)* ( 1 + k2 * tau))
    cc = (k1 * k2 * ca0 * tau**2) / ((1 + k1* tau)* ( 1 + k2 * tau))
    return [ca, cb, cc]

def update_plots_plotly(ca0, cb0, cc0, k1, k2):
    try:
        t_span = (0, 25)
        t_eval = np.arange(*t_span, 0.01)
        x0 = [ca0, cb0, cc0]
        k = [k1, k2]
        ca, cb, cc = model_cstr(t_eval, x0, k)
        
        max_cb = np.argmax(cb)

        fig = go.Figure()
        fig.add_trace(go.Scatter(x=t_eval, y=ca, mode='lines', name='Ca(t)'))
        fig.add_trace(go.Scatter(x=t_eval, y=cb, mode='lines', name='Cb(t)'))
        fig.add_trace(go.Scatter(x=t_eval, y=cc, mode='lines', name='Cc(t)'))
        
        fig.add_vline(x=t_eval[max_cb], line_dash="dash", line_color="red")
        
        annotation_text = (f'Max Cb: {cb[max_cb]:.2f} at tau={t_eval[max_cb]:.2f}<br>'+ # notar que el <br> es para salto de linea en plotly
                           f'Ca={ca[max_cb]:.2f}, Cb={cb[max_cb]:.2f}, Cc={cc[max_cb]:.2f}<br>'+
                           f'S_dc = {cb[max_cb]/cc[max_cb]:.2f}, Yb = {cb[max_cb]/(ca0 - ca[max_cb]):.2f}')
        
        fig.add_annotation(
            x=t_eval[max_cb],
            y=cb[max_cb] + 0.5,
            text=annotation_text,
            showarrow=False,
            xanchor="left",
            yanchor="bottom",
            bgcolor="white",
            bordercolor="black",
            font=dict(color="black", size=10)
        )
        
        fig.update_layout(
            title='CSTR Simulation (Solucion Analítica)',
            xaxis_title='Tiempo de contancto (tau)',
            yaxis_title='Concentration',
            width=800,
            height=500
        )
        fig.show(renderer="colab") # para comorerlo en propio usar fig.show()
    except Exception as e:
        print(f"Error: {e}")

# Create interactive widgets
ca0_slider_pl = widgets.FloatSlider(value=2, min=0.1, max=5, step=0.1, description='Ca0:')
cb0_slider_pl = widgets.FloatSlider(value=0, min=0.1, max=5, step=0.1, description='Cb0:')
cc0_slider_pl = widgets.FloatSlider(value=0, min=0.1, max=5, step=0.1, description='Cc0:')
k1_slider_pl = widgets.FloatSlider(value=0.5, min=1e-10, max=1, step=0.1, description='k1:')
k2_slider_pl = widgets.FloatSlider(value=0.2, min=1e-10, max=1, step=0.1, description='k2:')

widgets.interact(update_plots_plotly, ca0 = ca0_slider_pl, cb0 = cb0_slider_pl, cc0 = cc0_slider_pl, 
                 k1 = k1_slider_pl, k2 = k2_slider_pl)
```

# PFR

```python
def model_pfr(x, c0, k, tau):
    '''
    Generates the balance for the equations in the system
    4A + 6B -> 5C + 6D
    2B -> C + E
    C + 2E -> 2F
    note that 
    r1a/-4 = r1b/-6 = r1c/5 = r1d/6
    r2b/-2 = r2c = r2e
    r3c/-1 = r3e/-2 = r3f/2
    '''
    ca, cb, cc, cd, ce, cf = x
    ca0, cb0 = c0
    k1b, k2c, k3e = k 
    
    # Rate definitions
    r1b = - k1b * ca * cb ** 1.5 # B is consumed in reaction 1
    r2c = k2c * cb ** 2 # C is generated in reaction 2
    r3e = - k3e * cc * ce ** 2 # E is consumed in reaction 3, so its net formation rate must be negative
    
    # Species balances based on relative velocities
    ra = 4 / 6 * r1b
    rb = r1b - 2 * r2c
    rc = -5 / 6 * r1b + r2c + 1 / 2 * r3e
    rd = - r1b
    re = r2c + r3e
    rf = - 1 / 2 * r3e
    
    # f() = 0
    fa = ca0 - ca + ra * tau
    fb = cb0 - cb + rb * tau
    fc =  - cc + rc * tau
    fd =  - cd + rd * tau
    fe = - ce + re * tau
    ff =  - cf + rf * tau
    return [fa, fb, fc, fd, fe, ff]

def update_plots_pfr(ca0, cb0, k1b, k2c, k3e):
    try:
        c0 = [ca0, cb0]
        k = [k1b, k2c, k3e]
        x0_guess = [ca0, cb0, 0.1, 0.1, 0.1, 0.1]  # Initial guess for concentrations
        solutions = [scipy.optimize.fsolve(model_pfr, x0_guess, args=(c0, k, tau)) for tau in np.linspace(0, 30, 50)]
        ca, cb, cc, cd, ce, cf = zip(*solutions)
        fig = go.Figure()
        fig.add_trace(go.Scatter(x=np.linspace(0, 30, 10), y=ca, mode='lines+markers', name='Ca(tau)'))
        fig.add_trace(go.Scatter(x=np.linspace(0, 30, 10), y=cb, mode='lines+markers', name='Cb(tau)'))
        fig.add_trace(go.Scatter(x=np.linspace(0, 30, 10), y=cc, mode='lines+markers', name='Cc(tau)'))
        fig.add_trace(go.Scatter(x=np.linspace(0, 30, 10), y=cd, mode='lines+markers', name='Cd(tau)'))
        fig.add_trace(go.Scatter(x=np.linspace(0, 30, 10), y=ce, mode='lines+markers', name='Ce(tau)'))
        fig.add_trace(go.Scatter(x=np.linspace(0, 30, 10), y=cf, mode='lines+markers', name='Cf(tau)'))
        fig.update_layout(
            title='PFR Simulation (Solucion Analítica)',
            xaxis_title='Tiempo de contancto (tau)',
            yaxis_title='Concentration',
            width=800,
            height=500
        )
        fig.show(renderer="colab") # para comorerlo en propio usar fig.show()
    except Exception as e:
        print(f"Error: {e}")

ca0_slider_pfr = widgets.FloatSlider(value=0.2, min=0.1, max=5, step=0.1, description='Ca0:')
cb0_slider_pfr = widgets.FloatSlider(value=0.85, min=0.1, max=5, step=0.1, description='Cb0:')
k1b_slider_pfr = widgets.FloatSlider(value=0.64, min=1e-10, max=1, step=0.01, description='k1b:')
k2c_slider_pfr = widgets.FloatSlider(value=0.33, min=1e-10, max=1, step=0.01, description='k2c:')
k3e_slider_pfr = widgets.FloatSlider(value=0.15, min=1e-10, max=1, step=0.01, description='k3e:')

widgets.interact(update_plots_pfr, ca0 = ca0_slider_pfr, cb0 = cb0_slider_pfr, 
                 k1b = k1b_slider_pfr, k2c = k2c_slider_pfr, k3e = k3e_slider_pfr)
```
