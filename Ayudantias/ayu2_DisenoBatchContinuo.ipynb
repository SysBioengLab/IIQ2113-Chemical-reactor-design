```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import pandas as pd
```

# Gráficos de Levenspiel

```python
# problemas 2.1 y 2.2
def neg_ra(X, k, ca0):
    ra = k * ca0 * (1 - X)
    return ra

X_plot = np.linspace(0, 0.9, 100)
X = np.linspace(0, 0.8, 100)
k = 0.5
ca0 = 0.4
Fa0 = 0.4

n_ra_plot = neg_ra(X_plot, k, ca0)
n_ra_eval = neg_ra(X, k, ca0)
plt.plot(X_plot,Fa0/ n_ra_plot)
plt.plot(X,Fa0/ n_ra_eval, linestyle='dashed', linewidth = 3)

area_pfr = sp.integrate.trapezoid(Fa0/ n_ra_eval, X)
area_cstr = 0.8 * Fa0 / neg_ra(0.8, k, ca0)

plt.fill_between(X, Fa0/ n_ra_eval, alpha=0.5, label =f'Area PFR = {area_pfr:.2f} L')
plt.fill_between(X, 0, Fa0/ neg_ra(0.8, k, ca0), alpha=0.5, label =f'Area CSTR = {area_cstr:.2f} L')

plt.xlabel('X')
plt.ylabel('Fa0/ra (L)')
plt.legend()
```

```python
# problema 2.3
X_0 = 0
X_1 = 2/3
X_2 = 8/9

n_ra_1 = neg_ra(X_1, k, ca0)
n_ra_2 = neg_ra(X_2, k, ca0)
plt.plot(X_plot,Fa0/ n_ra_plot)
area1 = (X_1 - X_0) * Fa0 / n_ra_1
area2 = (X_2 - X_1) * Fa0 / n_ra_2

plt.fill_between(np.linspace(X_0, X_1, 100), Fa0/ n_ra_1, alpha=0.5, label =f'CSTR 1 = {area1:.2f} L')
plt.fill_between(np.linspace(X_1, X_2, 100), Fa0/ n_ra_2, alpha=0.5, label =f'CSTR 2 = {area2:.2f} L')

plt.xlabel('X')
plt.ylabel('Fa0/ra (L)')
plt.legend()
```

```python
# problema 2.4

X_0  = 0
X_1 = 1 - np.exp(-2)
X_2 = 1 - np.exp(-4)
n_ra_eval_1 = neg_ra(np.linspace(X_0, X_1, 100), k, ca0)
n_ra_eval_2 = neg_ra(np.linspace(X_1, X_2, 100), k, ca0)
n_ra_plot = neg_ra(np.linspace(X_0, X_2, 100), k, ca0)
plt.plot(np.linspace(X_0, X_2, 100), Fa0/ n_ra_plot)

area1 = sp.integrate.trapezoid(Fa0/ n_ra_eval_1, np.linspace(X_0, X_1, 100))
area2 = sp.integrate.trapezoid(Fa0/ n_ra_eval_2, np.linspace(X_1, X_2, 100))    

plt.fill_between(np.linspace(X_0, X_1, 100), Fa0/ n_ra_eval_1, alpha=0.5, label =f'PFR 1 = {area1:.2f} L')
plt.fill_between(np.linspace(X_1, X_2, 100), Fa0/ n_ra_eval_2, alpha=0.5, label =f'PFR 2 = {area2:.2f} L')
plt.xlabel('X')
plt.ylabel('Fa0/ra (L)')
plt.legend()
```

```python
# 2.7
X_0 = 0
X_1 = 1-np.exp(-2)
X_2 = (2 + 1 - np.exp(-2)) / 3

neg_ra_eval_1 = neg_ra(np.linspace(X_0, X_1, 100), k, ca0)

area1 = sp.integrate.trapezoid(Fa0/ neg_ra_eval_1, np.linspace(X_0, X_1, 100))
area2 = (X_2 - X_1) * Fa0 / neg_ra(X_2, k, ca0)

plt.plot(np.linspace(X_0, X_2, 100), Fa0 / neg_ra(np.linspace(X_0, X_2, 100), k, ca0))

plt.fill_between(np.linspace(X_0, X_1, 100), Fa0/ neg_ra_eval_1, alpha=0.5, label =f'PFR 1 = {area1:.2f} L')
plt.fill_between(np.linspace(X_1, X_2, 100), 0, Fa0/ neg_ra(X_2, k, ca0), alpha=0.5, label =f'CSTR 2 = {area2:.2f} L')
plt.axhline( Fa0 / neg_ra(X_2, k, ca0))
plt.xlabel('X')
plt.ylabel('Fa0/ra (L)')
plt.legend()
```

# Pregunta N-1

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
datos = {
    'X': [0, 0.2, 0.4, 0.6, 0.65],
    'neg_ra': [39, 53, 59, 38, 25]
}
# Transformamos el diccionario en un DataFrame
df = pd.DataFrame(datos)
# Calculamos el termino (F_A0 / -r_A) para el grafico
F_A0 = 50 # kmol/h
df['FA0_sobre_ra'] = F_A0 / df['neg_ra']
```

```python
# area CSTR 1 desde X_0 = 0 hasta X_1 = 0.2
area_cstr1 = (0.2 - 0) * df.loc[df['X'] == 0.2, 'FA0_sobre_ra'].values[0]
print(f'Area CSTR 1: {area_cstr1:.2f} L')

# area PFR desde X_1 = 0.2 hasta X_2 = 0.6

X_pfr = df.loc[(df['X'] >= 0.2) & (df['X'] <= 0.6), 'X']
Y_pfr = df.loc[(df['X'] >= 0.2) & (df['X'] <= 0.6), 'FA0_sobre_ra']
area_pfr = sp.integrate.trapezoid(Y_pfr, X_pfr)
print(f'Area PFR: {area_pfr:.2f} L')

# area CSTR 2 desde X_2 = 0.6 hasta X_3 = 0.65
area_cstr2 = (0.65 - 0.6) * df.loc[df['X'] == 0.65, 'FA0_sobre_ra'].values[0]
print(f'Area CSTR 2: {area_cstr2:.2f} L')
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(df['X'], df['FA0_sobre_ra'], marker='o', linestyle='-')
ax.fill_between(X_pfr, Y_pfr, alpha=0.5, label=f'Area PFR = {area_pfr:.2f} L')
ax.fill_between([0, 0.2], 0, df.loc[df['X'] == 0.2, 'FA0_sobre_ra'].values[0], alpha=0.5, label=f'Area CSTR 1 = {area_cstr1:.2f} L')
ax.fill_between([0.6, 0.65], 0, df.loc[df['X'] == 0.65, 'FA0_sobre_ra'].values[0], alpha=0.5, label=f'Area CSTR 2 = {area_cstr2:.2f} L')
ax.set_xlabel('X')
ax.set_ylabel('Fa0/-ra (L)')
ax.legend()
plt.savefig('data_plots/diseno_batch_continuo_1.pdf', transparent = True)
```
